## Testkorpus

Sampler noen verk fra tiårene 1970 til 2000. 300 fra dewey. Skjønnlitteratur fra korpusliste

```python
sakprosa = nb.pure_urn(wrd.get_text_from_word('ressurser/Korpus Avstandsblikk på norsk litteraturhistorie.docx'))
```

```python
sakprosa = nb.frame(sakprosa, 'urn')
sakprosa.head()
```

```python
romaner = pd.read_excel('ressurser/Romanoversikt frem til 1899 med grafer og greier.xlsx')
```

```python
romaner.head()
```

```python
romaner = romaner['URN']
```

```python
romaner = nb.frame([y[0] for y in [nb.pure_urn(x) for x in romaner] if y != []], 'urn')
```

Vi teller opp ordene i begge korpus

```python
sagprosa_agg = nb.aggregate_urns(list(sakprosa.urn))
```

```python
romaner_agg = nb.aggregate_urns(list(romaner.urn))
```

og legger dem i en dataramme

```python
combo = pd.concat([nb.frame(romaner_agg, 'skjønnlitt'), nb.frame(sagprosa_agg, 'sakprosa')], axis = 1, sort=False).dropna()
```

datarammen kan vi så sette opp for å sjekke forholdet mellom bruk av tegn og ord

```python
combo
```

Normaliserer for å se forholdstallene tydelig, så de ikke påvirkes av størrelsen på korpuset

```python
nb.normalize_corpus_dataframe(combo)
```

```python
combo['ratio'] = combo['skjønnlitt']/combo['sakprosa']
```
